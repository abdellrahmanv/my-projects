```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications import VGG16
import matplotlib.pyplot as plt
import numpy as np
import os
from collections import Counter
import seaborn as sns
import warnings

warnings.filterwarnings('ignore', category=UserWarning)

# Dataset path
data_dir = '/kaggle/input/brain-cancer-mri-dataset/Brain_Cancer raw MRI data/Brain_Cancer'

IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 8
NUM_CLASSES = 3

tf.random.set_seed(42)

def compute_mean_std(generator):
    mean, std = 0, 0
    n_pixels = 0
    generator.reset()
    for i in range(min(len(generator), 100)):
        images, _ = next(generator)
        batch_mean = np.mean(images, axis=(0, 1, 2))
        batch_std = np.std(images, axis=(0, 1, 2))
        batch_pixels = images.shape[0] * images.shape[1] * images.shape[2]
        mean += batch_mean * batch_pixels
        std += batch_std * batch_pixels
        n_pixels += batch_pixels
    mean /= n_pixels
    std /= n_pixels
    return mean, std

# Data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    zoom_range=0.4,
    shear_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

test_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

temp_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
temp_generator = temp_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    subset='training',
    shuffle=False
)
mean, std = compute_mean_std(temp_generator)
print(f"Dataset mean: {mean}, std: {std}")

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    subset='training',
    shuffle=True
)
test_generator = test_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    subset='validation',
    shuffle=False
)

# Validate generator
images, labels = next(train_generator)
print(f"Train batch shape: {images.shape}, Labels shape: {labels.shape}")
images, labels = next(test_generator)
print(f"Test batch shape: {images.shape}, Labels shape: {labels.shape}")

# Class weights
class_counts = {}
for cls in os.listdir(data_dir):
    class_path = os.path.join(data_dir, cls)
    if os.path.isdir(class_path):
        class_counts[train_generator.class_indices[cls]] = len(os.listdir(class_path))

total = sum(class_counts.values()) * 0.8
class_weights = {i: total/(NUM_CLASSES * count * 0.8) for i, count in class_counts.items()}
print("Class counts:", class_counts)
print("Class weights:", class_weights)

# Define VGG16 with transfer learning
def build_vgg16(input_shape=(224, 224, 3), num_classes=3):
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    base_model.trainable = False
    
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(256, activation='relu', kernel_regularizer=l2(1e-3)),
        BatchNormalization(),
        Dropout(0.6),
        Dense(num_classes, activation='softmax')
    ])
    return model

# Create and compile
model = build_vgg16(input_shape=(224, 224, 3), num_classes=3)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# Callbacks
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True, verbose=1)
lr_reducer = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=3, min_lr=1e-5, verbose=1)
checkpoint = ModelCheckpoint(
    '/kaggle/working/vgg16_transfer_best.keras',
    monitor='val_accuracy', save_best_only=True, mode='max', verbose=1
)

# Train
try:
    history = model.fit(
        train_generator,
        epochs=20,
        validation_data=test_generator,
        class_weight=class_weights,
        callbacks=[early_stopping, lr_reducer, checkpoint],
        verbose=1
    )
except Exception as e:
    print(f"Training failed: {e}")
    raise

# Fine-tune
model.layers[0].trainable = True
for layer in model.layers[0].layers[:-4]:
    layer.trainable = False

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

try:
    history_fine = model.fit(
        train_generator,
        epochs=10,
        validation_data=test_generator,
        class_weight=class_weights,
        callbacks=[early_stopping, lr_reducer, checkpoint],
        verbose=1
    )
except Exception as e:
    print(f"Fine-tuning failed: {e}")
    raise

# Evaluate
test_loss, test_accuracy = model.evaluate(test_generator, verbose=1)
print(f"Test Accuracy: {test_accuracy*100:.2f}%")

# Save model
try:
    model.save('/kaggle/working/vgg16_transfer_final.keras')
    print("Model saved as '/kaggle/working/vgg16_transfer_final.keras'")
except Exception as e:
    print(f"Failed to save .keras model: {e}")
    try:
        model.save('/kaggle/working/vgg16_transfer_final.h5')
        print("Model saved as '/kaggle/working/vgg16_transfer_final.h5'")
    except Exception as e2:
        print(f"Failed to save .h5 model: {e2}")

# Verify saved model
try:
    saved_model = tf.keras.models.load_model('/kaggle/working/vgg16_transfer_final.keras')
    saved_loss, saved_accuracy = saved_model.evaluate(test_generator, verbose=1)
    print(f"Saved Model Test Accuracy: {saved_accuracy*100:.2f}%")
except Exception as e:
    print(f"Failed to load/verify .keras model: {e}")
    try:
        saved_model = tf.keras.models.load_model('/kaggle/working/vgg16_transfer_final.h5')
        saved_loss, saved_accuracy = saved_model.evaluate(test_generator, verbose=1)
        print(f"Saved Model Test Accuracy: {saved_accuracy*100:.2f}%")
    except Exception as e2:
        print(f"Failed to load/verify .h5 model: {e2}")

# Visualize predictions
images_to_show = []
true_labels = []
pred_labels = []

test_generator.reset()
for i in range(30 // BATCH_SIZE + 1):
    images, labels = next(test_generator)
    predictions = model.predict(images, verbose=0)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(labels, axis=1)
    images_to_show.extend(images)
    true_labels.extend(true_classes)
    pred_labels.extend(predicted_classes)
    if len(images_to_show) >= 30:
        images_to_show = images_to_show[:30]
        true_labels = true_labels[:30]
        pred_labels = pred_labels[:30]
        break

class_names = list(train_generator.class_indices.keys())
plt.figure(figsize=(15, 18))
for i in range(30):
    plt.subplot(6, 5, i + 1)
    img = images_to_show[i][:, :, 0]
    plt.imshow(img, cmap='gray')
    plt.title(f"True: {class_names[true_labels[i]]}\nPred: {class_names[pred_labels[i]]}", fontsize=8)
    plt.axis('off')
plt.tight_layout()
plt.savefig('/kaggle/working/vgg16_transfer_predictions.png')
plt.show()

# Metrics
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix

test_generator.reset()
y_true = []
y_pred = []
for i in range(len(test_generator)):
    images, labels = next(test_generator)
    predictions = model.predict(images, verbose=0)
    y_true.extend(np.argmax(labels, axis=1))
    y_pred.extend(np.argmax(predictions, axis=1))

precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
print(f"Overall - Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")

for i, class_name in enumerate(class_names):
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average=None)
    print(f"{class_name} - Precision: {precision[i]:.4f}, Recall: {recall[i]:.4f}, F1-Score: {f1[i]:.4f}")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig('/kaggle/working/vgg16_transfer_confusion_matrix.png')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications import ResNet101
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
import warnings

# Suppress warnings
warnings.filterwarnings('ignore', category=UserWarning)

# Dataset path
data_dir = '/kaggle/input/brain-cancer-mri-dataset/Brain_Cancer raw MRI data/Brain_Cancer'

# Image parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 8
NUM_CLASSES = 3

# Set random seed
tf.random.set_seed(42)

# Compute dataset mean and std (RGB)
def compute_mean_std(generator):
    mean, std = np.zeros(3), np.zeros(3)
    n_pixels = 0
    generator.reset()
    for i in range(min(len(generator), 100)):
        images, _ = next(generator)
        if images.shape[-1] != 3:
            raise ValueError(f"Expected RGB images, got shape {images.shape}")
        batch_mean = np.mean(images, axis=(0, 1, 2))
        batch_std = np.std(images, axis=(0, 1, 2))
        batch_pixels = images.shape[0] * images.shape[1] * images.shape[2]
        mean += batch_mean * batch_pixels
        std += batch_std * batch_pixels
        n_pixels += batch_pixels
    mean /= n_pixels
    std /= n_pixels
    return mean, std

# Data generators (RGB)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    shear_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

test_datagen = ImageDataGenerator(
    rescale=1./255,
    featurewise_center=True,
    featurewise_std_normalization=True,
    validation_split=0.2
)

temp_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
temp_generator = temp_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',  # Native RGB
    subset='training',
    shuffle=False
)
mean, std = compute_mean_std(temp_generator)
print(f"RGB mean: {mean}, std: {std}")

train_datagen.mean = mean
train_datagen.std = std
test_datagen.mean = mean
test_datagen.std = std

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    subset='training',
    shuffle=True
)
test_generator = test_datagen.flow_from_directory(
    data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    subset='validation',
    shuffle=False
)

# Validate generator
images, labels = next(train_generator)
print(f"Train batch shape: {images.shape}, Labels shape: {labels.shape}")
images, labels = next(test_generator)
print(f"Test batch shape: {images.shape}, Labels shape: {labels.shape}")

# Class weights
class_counts = {}
for cls in os.listdir(data_dir):
    class_path = os.path.join(data_dir, cls)
    if os.path.isdir(class_path):
        class_counts[train_generator.class_indices[cls]] = len(os.listdir(class_path))

total = sum(class_counts.values()) * 0.8
class_weights = {i: total/(NUM_CLASSES * count * 0.8) for i, count in class_counts.items()}
class_weights[0] = 2.0  # Boost brain_glioma
print("Class counts:", class_counts)
print("Class weights:", class_weights)

# Define ResNet101 for RGB
def build_resnet101_rgb(input_shape=(224, 224, 3), num_classes=3):
    base_model = ResNet101(weights='imagenet', include_top=False, input_shape=input_shape)
    model = Sequential()
    model.add(base_model)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(256, activation='relu', kernel_regularizer=l2(1e-3)))
    model.add(BatchNormalization())
    model.add(Dropout(0.6))
    model.add(Dense(num_classes, activation='softmax'))
    for layer in base_model.layers:
        layer.trainable = False
    return model

# Create and compile
model = build_resnet101_rgb(input_shape=(224, 224, 3), num_classes=3)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# Callbacks
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True, verbose=1)
lr_reducer = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=3, min_lr=1e-5, verbose=1)
checkpoint = ModelCheckpoint(
    '/kaggle/working/resnet101_rgb_best.keras',
    monitor='val_accuracy', save_best_only=True, mode='max', verbose=1
)

# Train
try:
    history = model.fit(
        train_generator,
        epochs=20,
        validation_data=test_generator,
        class_weight=class_weights,
        callbacks=[early_stopping, lr_reducer, checkpoint],
        verbose=1
    )
except Exception as e:
    print(f"Training failed: {e}")
    raise

# Fine-tune
for layer in model.layers[0].layers[-30:]:  # Unfreeze more layers
    layer.trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),  # Increased for faster convergence
              loss='categorical_crossentropy',
              metrics=['accuracy'])

try:
    history_fine = model.fit(
        train_generator,
        epochs=15,
        validation_data=test_generator,
        class_weight=class_weights,
        callbacks=[early_stopping, lr_reducer, checkpoint],
        verbose=1
    )
except Exception as e:
    print(f"Fine-tuning failed: {e}")
    raise

# Evaluate
test_loss, test_accuracy = model.evaluate(test_generator, verbose=1)
print(f"Test Accuracy: {test_accuracy*100:.2f}%")

# Metrics
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix

test_generator.reset()
y_true = []
y_pred = []
for i in range(len(test_generator)):
    images, labels = next(test_generator)
    predictions = model.predict(images, verbose=0)
    y_true.extend(np.argmax(labels, axis=1))
    y_pred.extend(np.argmax(predictions, axis=1))

class_names = list(train_generator.class_indices.keys())
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
print(f"Overall - Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")

for i, class_name in enumerate(class_names):
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average=None)
    print(f"{class_name} - Precision: {precision[i]:.4f}, Recall: {recall[i]:.4f}, F1-Score: {f1[i]:.4f}")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig('/kaggle/working/resnet101_rgb_confusion_matrix.png')
plt.show()

# Visualize predictions
images_to_show = []
true_labels = []
pred_labels = []

test_generator.reset()
for i in range(30 // BATCH_SIZE + 1):
    images, labels = next(test_generator)
    predictions = model.predict(images, verbose=0)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(labels, axis=1)
    images_to_show.extend(images)
    true_labels.extend(true_classes)
    pred_labels.extend(predicted_classes)
    if len(images_to_show) >= 30:
        images_to_show = images_to_show[:30]
        true_labels = true_labels[:30]
        pred_labels = pred_labels[:30]
        break

plt.figure(figsize=(15, 18))
for i in range(30):
    plt.subplot(6, 5, i + 1)
    img = images_to_show[i]  # RGB images
    plt.imshow(img)
    plt.title(f"True: {class_names[true_labels[i]]}\nPred: {class_names[pred_labels[i]]}", fontsize=8)
    plt.axis('off')
plt.tight_layout()
plt.savefig('/kaggle/working/resnet101_rgb_predictions.png')
plt.show()

# Save model
try:
    model.save('/kaggle/working/resnet101_rgb_final.keras')
    print("Saved as '/kaggle/working/resnet101_rgb_final.keras'")
except Exception as e:
    print(f"Failed .keras: {e}")
    try:
        model.save('/kaggle/working/resnet101_rgb_final.h5')
        print("Saved as '/kaggle/working/resnet101_rgb_final.h5'")
    except Exception as e2:
        print(f"Failed .h5: {e2}")

# Verify saved model
try:
    saved_model = load_model('/kaggle/working/resnet101_rgb_final.keras')
    saved_loss, saved_accuracy = saved_model.evaluate(test_generator, verbose=1)
    print(f"Saved Model Test Accuracy: {saved_accuracy*100:.2f}%")
except Exception as e:
    print(f"Failed to load .keras: {e}")
    try:
        saved_model = load_model('/kaggle/working/resnet101_rgb_final.h5')
        saved_loss, saved_accuracy = saved_model.evaluate(test_generator, verbose=1)
        print(f"Saved Model Test Accuracy: {saved_accuracy*100:.2f}%")
    except Exception as e2:
        print(f"Failed to load .h5: {e2}")

# Check disk
!ls -lh /kaggle/working
!df -h /kaggle/working
```
